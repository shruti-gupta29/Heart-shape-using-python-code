# IPL Data Analysis 2022

```python
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go
import matplotlib.pyplot as plt
import seaborn as sns 
from datetime import datetime
```

```python
data=pd.read_csv("ipl_dataset.csv")
```

```python
data
```

# Number of matches won by each team in IPL 

```python
figure=px.bar(data,x=data['winner'],title="Number of matches won in IPL 2022")
figure.show()
```

# find out the player of the match

```python
figure=px.bar(data,x=data["man_of_the_match"],title="Most player of the match awards")
figure.show()
```

# Find out matches won in stadium

```python
figure=px.bar(data, x=data["stadium"],y=data["place"],color=data["place"],title="Matches won in stadium")
figure.show()
```

# IF Top scorer over team 1 and team 2

```python
figure=px.bar(data, x=data["team1_score"],y=data["team2_score"],title="Top scorer over team 1 and team 2")
figure.show()
```

# winner of the matches over toss winner

```python
sns.lineplot(x=data.index,y=data["toss_winner"],label="winner")
plt.title="Winner of the matches"
```
